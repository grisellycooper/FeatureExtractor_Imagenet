### Get Distance metric

##### Chi-squared distance is used to check the similarity among two fecture vectors. Metric Rank-n and MSE info is returned

```python
import numpy as np 
import csv
import pandas as pd
import matplotlib.pyplot as plt  
%matplotlib inline

from sklearn.metrics import classification_report, confusion_matrix, mean_squared_error
```

```python
def chiSquared(a,b):
    res = np.sqrt(0.5*np.sum((a-b)**2/(a+b+1e-6)))
    return res

def chiSquared2(a,b):
    #res = np.sqrt(0.5*np.sum((a-b)**2/(a+b+1e-6)))
    return 0.5*np.sum((a-b)**2/(a+b+1e-6))

#Input: csvfile // Output: string with results of MSE and Rank1
def getMetrics(csvfilename, rank):
    data = pd.read_csv(csvfilename)
    X = data.drop('class', axis=1)  
    y = data['class']
    X_np = X.values
    y_np = y.values
    s = (len(X_np), len(X_np), 2)
    y_pred = np.zeros(s)
    
    # Get distances matrix
    for i in range(len(X_np)):
        #minimun = 1e+10;    
        for j in range (len(X_np)):
            dis = chiSquared(X_np[i], X_np[j])
            #if dis < minimun and i != j:
            #minimun = res
            y_pred[i][j] = [dis, y[j]] 
                
    # Order matrix acordance to distance
    for i in range(len(y_pred)):
        b = y_pred[i]
        y_pred[i] = b[b[:,0].argsort()] 
    
    count = 0
    # Get Rank counting
    rankbins = np.zeros(rank+1)
    for i in range (len(y_pred)):
        for j in range (len(y_pred)):
            if y[i] == y_pred[i][j][1]:
                count += 1
                if j < rank:
                    rankbins[j] += 1
                else:
                    rankbins[rank] += 1
                break
    
    #Mean Square Error
    #msq = np.sqrt(mean_squared_error(y, y_pred))
    #res = 'MSE: '+ str(msq)    
    return (count,rankbins)
```

```python
rankbsif = getMetrics("calculated_features/awe-awet_lpq.csv", 10)
rankbsif
```
